### Importation tableau APL de la France

```python
apl_df = pd.read_excel(os.path.join(DATA_DIR,TABLEAU_APL_FILE), skiprows=7)
apl_df = apl_df.drop(apl_df.index[0])
```

```python
apl_df.head(5)
```

```python
results_df = pd.merge(apl_df, doctors_df[['Code commune INSEE',"Nom d'exercice"]],on='Code commune INSEE', how='left')
```

```python
results_df.head(5)
```

```python
liste_medecins_df = results_df.groupby('Code commune INSEE')["Nom d'exercice"].apply(list)
```

```python
liste_medecins_df.columns
```

```python
liste_medecins_df = liste_medecins_df.to_frame().reset_index() # On le convertit en DF
```

```python
results_df = pd.merge(apl_df, liste_medecins_df[['Code commune INSEE',"Nom d'exercice"]],on='Code commune INSEE', how='left')
```
