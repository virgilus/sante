### Fonction de conversion d'un fichier Excel en CSV

```python
import xlrd
import csv

def csv_from_excel(PathIn,PathOut,SheetName):
    wb = xlrd.open_workbook(PathIn)
    sh = wb.sheet_by_name(SheetName)
    your_csv_file = open(PathOut, 'w', encoding='utf-8') # Encodage en utf-8
    wr = csv.writer(your_csv_file, quoting=csv.QUOTE_ALL)

    for rownum in range(sh.nrows):
        wr.writerow(sh.row_values(rownum))

    your_csv_file.close()
```
